# Verileri kategori etmek için pandas kütüphanesininden cut fonksiyonunuda kullanabiliriz

```python
my_labels=[str(agg_df['AGE'].min())+"_18","19_23", "24_30" , "31_40" , "41_"+str(agg_df['AGE'].max())]
```

```python
my_labels
```

```python
bins=[0,18,23,30,40,66]
bins
```

```python
agg_df["age_cat"] = pd.cut(agg_df["AGE"], bins , labels=my_labels) 
```

```python
agg_df
```

```python
agg_df.values[0][2]
```

#List comprehension
### new_list = [expression for item in iterable if condition]


```python
agg_df["customers_level_based"] = [row[0].upper() + "_" + row[1].upper() + "_" + row[2].upper() + "_" + row[5].upper() for row in agg_df.values]
```

```python
agg_df
```

```python
agg_df = agg_df.groupby("customers_level_based").agg({"PRICE":"mean"}).reset_index()
```

```python
agg_df
```

```python
agg_df.shape
```

```python
agg_df["customers_level_based"].value_counts()
```

```python
agg_df["SEGMENT"] = pd.qcut(agg_df["PRICE"] ,  4 , labels=["A","B","C","D"])
```

```python
agg_df
```

```python
agg_df.groupby("SEGMENT").agg({"PRICE":"mean"})
```
